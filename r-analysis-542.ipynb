```r
data <- read.csv("/kaggle/input/data-clean/data_clean.csv")
```

```r
summary(data)
```

```r
library(MASS)
```

```r
attach(data)
```

```r
scope_data <- formula(Sleep.efficiency~Age+Gender+Sleep.duration+ REM.sleep.percentage+Deep.sleep.percentage+Light.sleep.percentage+Awakenings+    Caffeine.consumption+Alcohol.consumption+Smoking.status+ Exercise.frequency)
```

```r
start <- lm(Sleep.efficiency~1, data)
model_forward <- stepAIC(start, scope_data, direction="forward")
```

```r
full <- lm(Sleep.efficiency~Age+Gender+Sleep.duration+ REM.sleep.percentage+Deep.sleep.percentage+Light.sleep.percentage+Awakenings+    Caffeine.consumption+Alcohol.consumption+Smoking.status+ Exercise.frequency, data)
```

```r
model_backward <- stepAIC(full, scope_data, direction="backward")
```

```r
summary(model_forward)
```

```r
summary(model_backward)
```

```r
AIC(model_backward)
```

```r
AIC(model_forward)
```

```r
BIC(model_backward)
```

```r
BIC(model_forward)
```
